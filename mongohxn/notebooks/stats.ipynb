The script generates statistics for the filestore-new collections

```python
from datetime import datetime

import numpy as np
import matplotlib.pyplot as plt
import h5py

import pprint

import pymongo
from pymongo import MongoClient

# make graphics inline
%matplotlib inline
```

```python
client = MongoClient('xf03id1-mdb01', 27017)
```

## Filestore DB

```python
db = client['filestore-new']
db.collection_names()
```

```python
# collection: sentinel
sentinels = db['sentinel']
pprint.pprint(sentinels.find_one())
print(sentinels.count())
```
